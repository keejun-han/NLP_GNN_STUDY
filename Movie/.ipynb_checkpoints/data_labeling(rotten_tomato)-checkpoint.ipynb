### 2. Sample Testing

```python
df = pd.DataFrame({'score': [ '95','96/100','3.5/5.7','7.5/5','90/95', '0/4', '5/0', 'A+', 'B-', 'C']})
```

```python
import re
pattern1 = re.compile('[\d\.]*\/[\d\.]*') 
tt = df['score'].str.findall(pattern1).str.get(0)
```

```python
tt
```

```python
for row in df['score'].index:
    print(row)
```

```python
df
```

```python
df2 = df.copy()
```

```python
pattern1 = re.compile('([\d\.]*)\/([\d\.]*)') # 숫자 데이터
pattern2 = re.compile('(\w)') # 카테고리 데이터
delete_list = []

for row in df2['score'].index:
    print(row, df2.loc[row, 'score'])
    num_value = pattern1.search(df2.loc[row, 'score'])
    cat_value = pattern2.search(df2.loc[row, 'score'])

    # 숫자/숫자 형태
    if(num_value):
        print('num')
        numerator = float(num_value.group(1))
        denominator = float(num_value.group(2))
        # 분모가 0인 경우
        if(denominator == 0):
            delete_list.append(row)
            print("삭제대상")
        else:
            print(numerator/denominator)
            df2.loc[row, 'score'] = numerator/denominator
    
    # 카테고리 형태
    elif(cat_value):
        print('cat')
        print(cat_value.group(0))
        df2.loc[row, 'score'] = cat_value.group(0)
    
    print()
```

```python
delete_list
```

```python
df2.drop(df2.index[delete_list], inplace=True)
```

```python
df2
```

- Categorical 값 변환

```python
cat_score = {'A': 1, 'B': 0.7, 'C': 0.5, 'D': 0.3, 'F': 0}
df2.replace(cat_score, inplace=True)
```

### 3. Review score 적용
- 전처리 예상 시간: 3시간 소요

```python
df = rotten_review.copy()
```

```python
df.head()
```

```python
import time
start_time = time.time()

pattern1 = re.compile('([\d\.]*)\/([\d\.]*)') # 숫자 데이터
pattern2 = re.compile('(\w)') # 카테고리 데이터
delete_list = []

for row in df.index:
#     print(row, df.loc[row, 'review_score'])
    num_value = pattern1.search(df.loc[row, 'review_score'])
    cat_value = pattern2.search(df.loc[row, 'review_score'])

    # 숫자/숫자 형태
    if(num_value):
#         print('num')
        numerator = float(num_value.group(1))
        denominator = float(num_value.group(2))
        # 분모가 0인 경우
        if(denominator == 0):
            delete_list.append(row)
#             print("삭제대상")
        else:
#             print(numerator/denominator)
            df.loc[row, 'review_score'] = numerator/denominator
    
    # 카테고리 형태
    elif(cat_value):
#         print('cat')
#         print(cat_value.group(0))
        df.loc[row, 'review_score'] = cat_value.group(0)
    
#     print()
```

```python
# 시간 표시 함수
def format_time(elapsed):
    # 반올림
    elapsed_rounded = int(round((elapsed)))
    # hh:mm:ss으로 형태 변경
    return str(datetime.timedelta(seconds=elapsed_rounded))
```

```python
print("  Data Labeling took: {:}".format(format_time(time.time() - start_time)))
```

```python
# 분모가 0인 경우
delete_list
```

```python
df.drop(index = delete_list, inplace=True)
```

```python
cat_score = {'A': 1, 'B': 0.7, 'C': 0.5, 'D': 0.3, 'F': 0}
df.replace(cat_score, inplace=True)
```

```python
set(df.review_score)
```

```python
path
```

```python
# 1차 저장
df.to_csv(path+'rotten_review_scaled_new.csv', index=True)
```

### 5. Label 설정

```python
clean_df = pd.read_csv(path+'rotten_review_scaled_clean.csv')
clean_df.drop('Unnamed: 0', axis=1,inplace=True)
```

```python
clean_df.head()
```

- review_score >= 0.7 : 1
- review_score <= 0.3 : 0

```python
pos_condition = clean_df.review_score >= 0.7
neg_condition = clean_df.review_score <= 0.3
```

```python
pos_list = clean_df[pos_condition].index
neg_list = clean_df[neg_condition].index
```

```python
clean_df.loc[pos_list, 'label'] = 1
clean_df.loc[neg_list, 'label'] = 0
```

```python
clean_df
```

```python
clean_df.dropna(axis=0, inplace=True)
```

```python
clean_df
```

```python
clean_df.label = clean_df.label.astype(int)
```

```python
clean_df.head()
```

```python
total_len = len(clean_df)
```

```python
print(f"Positive label 개수 : {sum(clean_df['label'] == 1.0)}")
print(f"Negative label 개수 : {sum(clean_df['label'] == 0.0)}")
```

```python
print(f"Positive label 비율 : {sum(clean_df['label'] == 1.0)/total_len}")
print(f"Negative label 비율 : {sum(clean_df['label'] == 0.0)/total_len}")
```

```python
clean_df.to_csv(path+'rotten_review_scaled_label.csv', index=False)
```
